```python
def get_team_valuation(team_name, team_valuations_df):
    """
    Finds the team valuation for a given team name (abbreviation or full name).

    Args:
        team_name (str): The name of the team to search for.
        team_valuations_df (pd.DataFrame): The DataFrame containing team valuations.

    Returns:
        float: The team valuation in millions of dollars, or None if not found.
    """
    try:
        # If team_name is an abbreviation, convert it to full name using the mapping
        full_team_name = team_name_mapping.get(team_name, team_name)

        # Find the row where the team name matches
        team_row = team_valuations_df[team_valuations_df['TEAM'] == full_team_name]
        # Extract the valuation from the 'VALUE_MILLIONS' column
        valuation = team_row['VALUE_MILLIONS'].iloc[0]
        return valuation
    except (IndexError, KeyError):
        # Return None if the team name is not found or the column is missing
        return None
# Apply the get_team_valuation function to add the Team_Valuation column
team_data['Team_Valuation'] = team_data['Team'].apply(lambda team: get_team_valuation(team, team_valuations))
# Display the result
print(team_data)
```

```python
# prompt: given database from team_data, give me Bar Chart Comparing Twitter Engagement, highlight BOS team another  Salary Across All Teams highlight BOS team as well with number on top of the bar

import plotly.express as px

# Assuming 'team_data' DataFrame is available and contains 'Team', 'Total Team Engagement', and 'Average Salary' columns.

fig = px.bar(team_data, x='Team', y='Total Team Engagement',
             title='Comparison of Twitter Engagement Across Teams',
             labels={'Team': 'Team Name', 'Total Team Engagement': 'Total Twitter Engagement'})

# Highlight BOS team in the Twitter Engagement chart
fig.update_traces(marker_color=['red' if team == 'BOS' else 'blue' for team in team_data['Team']])
bos_engagement = team_data[team_data['Team'] == 'BOS']['Total Team Engagement'].iloc[0]
fig.add_annotation(
    x='BOS',
    y=bos_engagement,
    text=str(round(bos_engagement,2)),
    showarrow=True,
    arrowhead=1
)

fig.show()


fig_salary = px.bar(team_data, x='Team', y='Average Salary',
                     title='Comparison of Average Salaries Across Teams',
                     labels={'Team': 'Team Name', 'Average Salary': 'Average Player Salary'})

# Highlight BOS team in the Salary chart
fig_salary.update_traces(marker_color=['red' if team == 'BOS' else 'blue' for team in team_data['Team']])

bos_salary = team_data[team_data['Team'] == 'BOS']['Average Salary'].iloc[0]
fig_salary.add_annotation(
    x='BOS',
    y=bos_salary,
    text=str(round(bos_salary,2)),
    showarrow=True,
    arrowhead=1
)
fig_salary.show()
```
